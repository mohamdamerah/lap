```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
#function for trapozidal
def trapmemberf(x,a,b,c,d):
    if (x <=a ):
        m=0
    elif (x> a and x<b):
        m=(x-a)/(b-a)
       
    elif (x>=b and x<=c):
        m=1
    elif (x>c and x<d):
        m=(d-x)/(d-c)
    elif(x>=d):
        m=0
        
    return m
```

```python
#thats for triangle 
def tramfun(x,y,l,n):
    if(x < y):
        s=0
    elif(x>=y and x<=l):
        s=(x-y)/(l-y)
    elif(x>=l and x<=n):
        s=(n-x)/(n-l)
    elif(x>n):
        s=0
    return s
```
